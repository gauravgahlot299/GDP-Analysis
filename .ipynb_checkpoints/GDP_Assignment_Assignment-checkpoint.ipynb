The data has data values for various states' GSDP in Crores and their percentage growth from previous years.

Task - Calculate the average growth of states for the duration 2013-14, 2014-15 and 2015-16 by taking the mean of the row 
 '(% Growth over previous year)'.

**Data Preparation 1** - Taking only data with 'Items Description' as '(% Growth over previous year)'.

```python
growth_percentage = gdpData.loc[gdpData['Items  Description'] == '(% Growth over previous year)']
growth_percentage
```

**Data Preparation 2** -
The 'Items Description' column has redundant values; therefore dropping it.
Also, 'Duration' column form a unique column. Opting for making it the index for this dataframe.

```python
growth_percentage = growth_percentage.drop(columns = ['Items  Description'])
growth_percentage.set_index('Duration', inplace = True)
print(growth_percentage.shape)
growth_percentage
```

**Data Cleaning 1** - It seems few 'NA' (null values) in the data for the state, choosing to ignore the NA values as considering them might create bias and can lead to wrong analysis since this is a very small dataset. 

**Data Cleaning 2** - Dropping the columns with all 'NA' values (like West Bengal).

```python
growth_percentage = growth_percentage.dropna(axis = 1,how = 'all')
print(growth_percentage.shape)
growth_percentage
```

**Data Preparation 3** - Taking data only for the duration 2013-14, 2014-15 and 2015-16.

```python
growth_percentage = growth_percentage.loc[growth_percentage.index.isin(["2013-14","2014-15","2015-16"])]
print(growth_percentage.shape)
growth_percentage.head()
```

**Data Calculation**
Taking the mean % Growth for every state

```python
avgGrowth = round(growth_percentage.mean(),2).sort_values(ascending = False)
avgGrowth
```

Task - Plot the mean % Growth for the states

**Data Visualization** - Creating bar plot with annotations and gradient coloring for better understanding.

```python
plt.figure(figsize=(18,6))
plt.xticks(rotation = 90)
clist = [(0, "red"), (0.125, "red"), (0.25, "orange"), (0.5, "green"), (0.7, "green"), (0.75, "blue"), (1, "blue")]
rvb = mcolors.LinearSegmentedColormap.from_list("", clist)
ax = avgGrowth.plot(kind='bar', color=rvb(avgGrowth/len(avgGrowth)), width=0.9)
for p in ax.patches: 
    ax.annotate(np.round(p.get_height(),1), (p.get_x()+p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title("Average Growth Rate for the duration 2013-14, 2014-15 and 2015-16, by State")
plt.ylabel('Average Growth (%)')
```

**From the above bar graph for Average Growth Rate for the duration 2013-14, 2014-15 and 2015-16, by State, we can infer that states of Mizoram, Tripura and Nagaland have been growing consistently fast, while Goa and Meghalaya have been struggling.

**Data Preparation 1 (Data-I-B)** -  Loading all the GSVA files for the states to separate data frames.

```python
files = glob.glob('./*.xlsx') #all files in current folder
dfs = {}

for f in files:
    filename = os.path.basename(f)
    if filename.startswith("NAD"):
        filename = filename.split("-")[1]
        print(filename)
        dfs[filename] = pd.read_excel(f)
```

Task - As is required to provide analysis only for the duration 2014-15, fetching only that from Data-I-B

**Data Preparation 2 (Data-I-B)** -

    Step 1 - Taking data for duration 2014-15 only
    Step 2 - Renaming the '2014-15' column to the State name (will be helpful in joining with other States' data later)
    Step 3 - Stripping of trailing asterisks(*) if there in Item column
    Step 4 - Setting Item column as the index of the dataframe

```python
for df in dfs:
    dfdata = dfs[df].loc[:,['Item','2014-15']]
    dfdata.rename(columns = {'2014-15':df},inplace = True)
    dfdata['Item'] = dfs[df]['Item'].str.rstrip('*')
    dfdata.set_index(['Item'], inplace = True)
    dfs[df] = dfdata
```

**Data Preparation 3 (Data-I-B)** -  Merging all the GSVA files i.e all the loaded data frames.

```python
Data_I_B = pd.DataFrame()
for df in dfs.values():
    Data_I_B = Data_I_B.join(df, how = "outer")
Data_I_B
```

Task - Filter out the union territories (Delhi, Chandigarh, Andaman and Nicobar Islands, etc.) for further analysis, as they are governed directly by the centre, not state governments.

**Data Preparation 4 (Data-I-B)** -  Removing data for UTs.

```python
Data_I_B = Data_I_B.loc[:,~Data_I_B.columns.isin(['Delhi','Chandigarh','Andaman_Nicobar_Islands','Puducherry'])]
print(Data_I_B.shape)                        
Data_I_B.head()
```

**Data Cleaning 1 (Data-I-B)** - There are few '0' (zeroes) in the data for the states. Approach used for handling this is replace all '0' with 'NA'.
Also, after this, I have decided **not to replace 'NA'** with any mean as this would not be good and lead to wrong data analysis.

```python
Data_I_B = Data_I_B.replace(0, np.nan)
Data_I_B.head()
```

Task - Plot the percentage contribution of the primary, secondary and tertiary sectors as a percentage of the total GDP for all the states

**Data Visualization 2**

**----------------State-wise----------------**
*Showing from state with highest GSDP to the state with lowest GSDP*

```python
srt = Data_I_B.sort_values(['Gross State Domestic Product'], axis = 1 ,ascending = False) #sorting the data as per GSDP
gdp_percentage = srt.loc[['Primary','Secondary','Tertiary']].div(srt.loc['Gross State Domestic Product']).mul(100)
data = gdp_percentage.values.tolist()
numberOfBars = len(data[0])
ind = list(range(numberOfBars))
axes = []
column_size = np.zeros(numberOfBars)
data = np.array(data)
segment_labels = gdp_percentage.index.tolist()
sector_labels = gdp_percentage.columns.tolist()
plt.figure(figsize=(15,6))
color =["#3778bf", '#feb308', '#a8a495']
for i, state_data in enumerate(data):
    axes.append(plt.bar(ind, state_data, bottom=column_size, label=segment_labels[i], color = color[i], width=0.9))
    column_size += state_data
plt.xticks(ind, sector_labels, rotation = 90)
plt.ylabel('Percentage (%)')
plt.xlabel('State')
plt.title('Percentage contribution of the primary, secondary and tertiary sectors as a percentage of the total GDP for all the states (State-wise)')
plt.legend()
for axis in axes:
    for bar in axis:
        w, h = bar.get_width(), bar.get_height()
        plt.text(bar.get_x() + w/2, bar.get_y() + h/2, "{:.1f}".format(h), ha="center", va="center")
```

**Inference** - For most of the Indian states, Tertiary Sector forms about 50% of their GDP except for few outliers like Uttarakhand and Sikkim where Secondary Sector contributes to about 50% of the GDP.
The top states in terms of GDP like Maharashtra, Tamil Nadu should continue investing in Tertiary sector, but are poor at the Primary ones. They should make a bit more investment in it. Also, if we use inference from Average Growth Percentage of these states, they have been at the medium pace, so if the geographic and demographics of the states are fortunate, the states can try seeing some good peak in growth rate if they experiment investing in the Crops, Forestry, etc. as well.
States of Sikkim and Uttarakhand have huge potential of Tertiary Sector like Tourism; they should invest in it also. These states and Goa are very poor at Primary Sector, they can try uplifting it via investing in Livestock, Fishing, Aquaculture. 

Task - Categorise the states into four groups based on the GDP per capita (C1, C2, C3, C4 where C1 would have the highest per capita GDP and C4, the lowest). The quantile values are (0.20,0.5, 0.85, 1), i.e., the states lying between the 85th and the 100th percentile are in C1; those between the 50th and the 85th percentiles are in C2, and so on.

**Data Manipulation** - Tranposing the data 'Data-I-B' and creating a new column named 'Category' which is outcome of the categorization of data (C1, C2, C3, C4) based on given quantiles.

```python
Data_I_B_Transposed = Data_I_B.T
quantiles = sorted(Data_I_B_Transposed['Per Capita GSDP (Rs.)'].quantile([0.20,0.5, 0.85, 1]).tolist(), reverse = True)
print('quantiles:',quantiles)

def categorize(x):
    if quantiles[1]<x<=quantiles[0]:
        return 'C1'
    elif quantiles[2]<x<=quantiles[1]:
        return 'C2'
    elif quantiles[3]<x<=quantiles[2]:
        return 'C3'
    else:
        return 'C4'
Data_I_B_Transposed['Category'] = Data_I_B_Transposed['Per Capita GSDP (Rs.)'].apply(lambda x : categorize(x))
Data_I_B_Transposed['Category'].head()
```

### Analysis on Category 'C1'

Task - Find the top 3/4/5 sub-sectors that contribute to approximately 80% of the GSDP of 'C1'.

**Data Preparation 1** - Subsetting the data frame to have only data values for 'C1' category and also, has information of the sub-sectors and GSDP only

```python
cols_reqd = ['Agriculture, forestry and fishing',
                'Mining and quarrying',
                'Manufacturing',
                'Electricity, gas, water supply & other utility services',
                'Construction',
                'Trade, repair, hotels and restaurants',
                'Transport, storage, communication & services related to broadcasting',
                'Financial services',
                'Real estate, ownership of dwelling & professional services',
                'Public administration',
                'Other services',
               'Gross State Domestic Product'
                ]
C1 = Data_I_B_Transposed.loc[Data_I_B_Transposed.Category == 'C1', cols_reqd]
C1 = C1.T
C1
```

**Data Calculation 1** - Finding the percentage contribution (to total GSDP of all the states ) of each sub-sector across the states in 'C1'.

```python
C1_new = C1.loc[:]
C1_new['Total'] = C1.sum(axis = 1) #temporary column for finding the total GSDP
C1_new['% of Total GSDP'] = C1_new['Total']/C1_new.loc['Gross State Domestic Product','Total']*100
C1_new = C1_new.drop('Gross State Domestic Product')
C1_new.drop('Total', axis = 1, inplace = True)
C1_new = C1_new.sort_values('% of Total GSDP', ascending = False)
C1_new
```

**Data Calculation 2** - Creating a column 'GSDP_Cumulative' which will have values of percentages uptil a particular sub-sector of C1 category.

```python
C1_new['GSDP_Cumulative'] = C1_new.loc[:,'% of Total GSDP'].cumsum()
C1_new
```

**Data Calculation 3** - Finding top 3/4/5 subsectors based on a column 'GSDP_Cumulative' of C1 category.

```python
print('Top 3 sub-sectors are -')
for i in C1_new.index[:3].tolist():
    print(i)
print('\nTop 4 sub-sectors are -')
for i in C1_new.index[:4].tolist():
    print(i)
print('\nTop 5 sub-sectors are -')
for i in C1_new.index[:5].tolist():
    print(i)
```

Task - Plot the contribution of the sub-sectors as a percentage of the GSDP of C1 category (sub-sectors making 80%)

**Data Visualization 1** - Choosing the bar chart for Data Visualization.

```python
top7 = C1_new.head(7)
sns.barplot(x = top7['% of Total GSDP'], y = top7.index)
plt.ylabel('Sub-Sectors')
plt.title("Contribution of the sub-sectors as a percentage of the GSDP of C1 category (sub-sectors making 80%)")
```

**Inference** - In case of C1 category states, Tertiary and Secondary Sectors are more accounting to the GDP. We can also see Agriculture (which is a Primary Sector) being among the top two. We see Manufacturing to be boon for C1.
These states can continue investing in real estate, agriculture, trade and manufacturing; but should also try making small investments in construction and transport.

## Part-II: GDP and Education Dropout Rates

Task - Analyse if there is any correlation of GDP per capita with dropout rates in education (primary, upper primary and secondary) for the year 2014-2015 for each state.

**Business Understanding** - 
Investigate whether there is any relationship between per capita GDP with dropout rates in education.

**Data Understanding** -

```python
# Loading the data (Data-II)

DropOut = pd.read_excel('rs_session243_au570_1.1.xlsx') #all files in the same folder
print(DropOut.shape)
print(DropOut.info())
DropOut
```

This data has the drop out rates in various states of Indian constituency from different levels of education over various durations.

**Data Preparation 1** - As is mentioned to consider the duration 2014-15 only and that too for only 3 levels of education (Primary, Upper Primary and Secondary), sub-setting the data.

```python
DropOut = DropOut.loc[:,['Level of Education - State', 'Primary - 2014-2015', 'Upper Primary - 2014-2015', 'Secondary - 2014-2015']]
print(DropOut.shape)
DropOut
```

**Data Preparation 2** - 

    Step 1 - Renaming the columns to more convenient names.
    Step 2 - Making the states as the index of the dataframe.
    Step 3 - Changing 'Chhatisgarh' to 'Chhattisgarh','Uttrakhand' to 'Uttarakhand' and 'Jammu and Kashmir' to 'Jammu Kashmir'.

```python
DropOut.rename(columns = {'Level of Education - State':'State', 
                          'Primary - 2014-2015':'Primary',
                          'Upper Primary - 2014-2015':'Upper Primary',
                          'Secondary - 2014-2015':'Secondary'}, inplace = True)
DropOut.set_index('State', inplace = True)
DropOut.rename(index = {'Chhatisgarh':'Chhattisgarh','Uttrakhand':'Uttarakhand', 'Jammu and Kashmir':'Jammu Kashmir'}, inplace = True)
DropOut
```

**Data Preparation 3** - Preparing Data-I-B-Transposed for merging with DropOut data.

    Step 1 - Taking only 'Per Capita GSDP (Rs.)' column.
    Step 2 - Renaming index to 'State'.
    Step 3 - Removing '_' from index.

```python
Data_I_B_1 = Data_I_B_Transposed[:]
Data_I_B_1 = Data_I_B_1.loc[:,['Per Capita GSDP (Rs.)']]
Data_I_B_1.index.names = ['State']
Data_I_B_1.index = Data_I_B_1.index.str.replace("_"," ")
Data_I_B_1
```

**Data Preparation 4** - Merging the dataframes 'DropOut' and 'Data_I_B_1'

```python
Merged = DropOut.merge(Data_I_B_1, how = 'inner', on = 'State')
Merged
```

**Data Calculation** - Finding correlations of various levels with 'Per Capita GSDP (Rs.)'

```python
print('With Primary:', Merged['Primary'].corr(Merged['Per Capita GSDP (Rs.)']))
print('With Upper Primary:', Merged['Upper Primary'].corr(Merged['Per Capita GSDP (Rs.)']))
print('With Secondary:', Merged['Secondary'].corr(Merged['Per Capita GSDP (Rs.)']))
```

Task - Choose an appropriate plot to conduct this analysis.

**Data Visualization** - Plotting scatter plots to conduct this analysis.

```python
sns.regplot(x = Merged['Per Capita GSDP (Rs.)'], y = Merged['Primary'])
plt.title("Correlation plot between Per Capita GSDP (Rs.) and Primary DropOut Rate")
plt.ylabel('Primary DropOut Rate (%)')
```

**Inference** - Per Capita GSDP (Rs.) and Primary DropOut Rate are negatively correlated. We can see that the states with low Per Capita GSDP have witnessed high droputs from the Primary education, and the vice-versa. The regression line is falling and states the same.

```python
sns.regplot(x = Merged['Per Capita GSDP (Rs.)'], y = Merged['Upper Primary'], color = '#a98d19')
plt.title("Correlation plot between Per Capita GSDP (Rs.) and Upper Primary DropOut Rate")
plt.ylabel('Upper Primary DropOut Rate (%)')
```

**Inference** - Per Capita GSDP (Rs.) and Upper Primary DropOut Rate are negatively correlated. We can see that the states with low Per Capita GSDP have witnessed high droputs from the Upper Primary education, and the vice-versa. The regression line is falling and states the same.

```python
sns.regplot(x = Merged['Per Capita GSDP (Rs.)'], y = Merged['Secondary'], color = '#a93f79')
plt.title("Correlation plot between Per Capita GSDP (Rs.) and Secondary DropOut Rate")
plt.ylabel('Secondary DropOut Rate (%)')
```

**Inference** - Per Capita GSDP (Rs.) and Secondary DropOut Rate are negatively correlated. We can see that the states with low Per Capita GSDP have witnessed high droputs from the Secondary education, and the vice-versa. The regression line is falling and states the same.
